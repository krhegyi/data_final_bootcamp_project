```python
#dropping unnecesarry columns
df_ml = df.drop(columns=['airline.iataCode',
 'airline.icaoCode',
 'Airline.Name',
 'Airline1',
 'Airline1 (group)',
 'arrival.actualTime',
 'arrival.iataCode',
 'arrival.scheduledTime',
 'biggest_EU_airlines',
 'departure.actualTime',
 'departure.scheduledTime',
 'flight_count',
 'Price ranking group',
 'Route Name',
 'Number_of_flights'])
```

```python
#oversampling dataset
from imblearn.over_sampling import SMOTE
sm = SMOTE()
```

```python
X_res, y_res = sm.fit_resample(df_ml.drop(columns=['delay_in_arrival_no delay']), df_ml['delay_in_arrival_no delay'])
```

```python
df_ml = pd.concat([X_res, y_res], axis=1)
```

```python
df_ml['delay_in_arrival_no delay'].value_counts()
```

# Machine Learning /Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X = sc.fit_transform(df_ml.drop('delay_in_arrival_no delay', axis=1))
y = df_ml['delay_in_arrival_no delay']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import plot_confusion_matrix, classification_report, accuracy_score
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
pred = rfc.predict(X_test)
print(classification_report(y_test, pred))
print(accuracy_score(y_test, pred))
```

```python
disp = plot_confusion_matrix(rfc, X_test, y_test,
                     display_labels= ["Delayed", 'Not delayed'])

disp.ax_.set_title("Confusion Matrix \n Random Forest")
```
