# Collect data from API -historical flights

```python
#airport list from the scraped airport (IATA) list notebook
airportsEU = ['AAL', 'ABZ', 'MAD', 'ALC', 'AMS', 'BCN', 'BRI', 'BFS', 'BEG',
       'BGO', 'SXF', 'TXL', 'BLL', 'BHX', 'BOO', 'BLQ', 'BOD', 'KBP',
       'BOH', 'BRE', 'BRS', 'BRU', 'CRL', 'BUD', 'BOJ', 'CAG', 'CWL',
       'CTA', 'CDG', 'CIA', 'CGN', 'CPH', 'ORK', 'DME', 'DTM', 'WRO',
       'DRS', 'DUB', 'DUS', 'EMA', 'EDI', 'EIN', 'ATH', 'BSL', 'EXT',
       'PMO', 'FAO', 'OPO', 'FRA', 'GDN', 'GVA', 'GOA', 'BHD', 'GLA',
       'GOT', 'LPA', 'GRV', 'HAM', 'HAJ', 'HEL', 'OTP', 'HER', 'LIS',
        'BGY', 'PDL', 'FKB', 'KTW', 'KZN', 'KEF', 'HRK', 'KRK', 'KUF',
        'SPC', 'TER', 'LBA', 'LEJ', 'TLL', 'FCO', 'LGG', 'LPL', 'LJU',
        'LGW', 'LHR', 'LTN', 'STN', 'LLA', 'LUX', 'LYS', 'BTS', 'AGP',
        'MMX', 'MXP', 'MLA', 'MAN', 'MRS', 'LIN', 'MSQ', 'WMI', 'MUC',
        'FMO', 'NAP', 'NCL', 'NCE', 'NWI', 'NUE', 'ODS', 'OSL', 'PMI',
        'ORY', 'PSA', 'ROV', 'TGD', 'POZ', 'PRN', 'LED', 'AKT', 'BZZ',
        'FFD', 'LKZ', 'MHZ', 'RMS', 'RIX', 'DSA', 'SCQ', 'SJJ', 'SNN',
        'SVO', 'SIP', 'SKP', 'AER', 'SOF', 'SOU', 'SVG', 'ARN', 'STR',
        'TFN', 'TFS', 'SKG', 'TIA', 'TLS', 'TSF', 'TOS', 'TRD', 'TRN',
        'UFA', 'PRG', 'VAR', 'VCE', 'VRN', 'VIE', 'VNO', 'VKO', 'WAW',
        'KJA', 'ZAG', 'ZIA', 'ZRH']
 
airportsDE = ['SXF','TXL','BRE','CGN','DTM','DRS','DUS','FRA','HAM',
               'HAJ','FKB','LEJ','MUC','FMO','NUE','STR'] #excluding RMS as it is a military airport


responses = []
 
startdate = pd.to_datetime("2020-06-01")
 
for x in airportsDE:
    while startdate < pd.to_datetime("2020-08-30"):
        enddate = startdate + pd.DateOffset(days=10)
        while True:
            try:
                url = 'http://aviation-edge.com/v2/public/flightsHistory?key=21aad3-65ccaa&code=' + x + '&type=arrival&status=landed&date_from=' + startdate.strftime('%Y-%m-%d') + '&date_to=' + enddate.strftime('%Y-%m-%d')
                response = requests.get(url)
                flights = [flight for flight in response.json() if not isinstance(flight, str)]
                responses += flights
                print(x, startdate, enddate, "Flights: ", len(flights))
            except JSONDecodeError:
                print("JSON decode error, retrying...")
                continue
            break
        startdate = enddate
    startdate = pd.to_datetime("2020-06-01")


print("All flights found: ", len(responses))
```

```python
pd.json_normalize(responses)
```

```python
pd.set_option('display.max_columns', None)
df_flights = pd.DataFrame(pd.json_normalize(responses))
df_flights.head()
```

# Data cleaning

```python
#IATA code to uppercase
df_flights['departure.iataCode'] = df_flights['departure.iataCode'].str.upper()
df_flights['arrival.iataCode'] = df_flights['arrival.iataCode'].str.upper()
df_flights.head(10)
```

```python
#drop unnecessary columns

df_flights.drop(df_flights.columns[[0, 1, 3, 4, 5, 8, 10, 11, 13, 14, 16, 18, 19, 23, 
                                    24, 25, 27, 28, 29, 30, 31, 32, 33, 34]], axis=1, inplace=True)
```

```python
#drop rows which are not in the airportEU list
df_flights = df_flights[df_flights['arrival.iataCode'].isin(airportsEU)]
df_flights = df_flights[df_flights['departure.iataCode'].isin(airportsEU)]
```

```python
#converting to datetime
df_flights['departure.scheduledTime']= pd.to_datetime(df_flights['departure.scheduledTime'])
df_flights['departure.actualTime']= pd.to_datetime(df_flights['departure.actualTime'])
df_flights['arrival.scheduledTime']= pd.to_datetime(df_flights['arrival.scheduledTime'])
df_flights['arrival.actualTime']= pd.to_datetime(df_flights['arrival.actualTime'])
```

```python
#delete rows with null values in any of the datetime column
df_flights = df_flights[df_flights['departure.actualTime'].notna()]
```
