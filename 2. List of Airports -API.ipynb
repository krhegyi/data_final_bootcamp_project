The aim of this notebook is to get a list of airports with their latitude/longitude data

```python
import requests
import pandas as pd
import json
from json.decoder import JSONDecodeError
```

```python
#list of airports
url = 'http://aviation-edge.com/v2/public/airportDatabase?key=21aad3-65ccaa'
response = requests.get(url)
response.json()
```

```python
pd.json_normalize(response.json())
```

```python
pd.set_option('display.max_columns', None)
df_airports = pd.DataFrame(pd.json_normalize(response.json()))
```

```python
df_airports.drop(df_airports.columns[[4,5,6,10,11,12]], axis=1, inplace=True)
```
